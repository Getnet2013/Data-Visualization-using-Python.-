```python
#introducing multiple Bar Chart Parameter
#modified (names onx-axis)
#Rotation Parameter
#Arange Function
import numpy as np
x=["1980","1981","1982","1983","1984","1985"]
#define the turnover values for company 1 and 2
y=[38000,45000,48000,52000,58000,62000]
z=[30000,40000,45000,50000,55000,60000]
width=0.4
p=np.arange(len(x))
p1=[j+width for j in p]
#Create the Bar Chart
plt.bar(x,y,width=0.6,color="r",edgecolor="black",label="company 1")
plt.bar(x,z,width=0.6,color="y",edgecolor="black",label="company 2")

#Add Label and Legend
plt.xlabel("Year")
plt.ylabel("turnover")
plt.title("Turnover ot two Companies")
plt.xticks(p+width/2,x,rotation=20)
plt.legend()

#show the  bar chart
plt.show()
```

```python
#Introducing Horizontal Bar Chart
import numpy as np
x=["1980","1981","1982","1983","1984","1985"]
#define the turnover values for company 1 and 2
y=[38000,45000,48000,52000,58000,62000]
z=[30000,40000,45000,50000,55000,60000]
width=0.4
p=np.arange(len(x))
p1=[j+width for j in p]
#Create the Bar Chart
plt.barh(p,y,width,color="r",edgecolor="black",label="company 1")
plt.barh(p1,z,width,color="y",edgecolor="black",label="company 2")

#Add Label and Legend
plt.xlabel("Year")
plt.ylabel("turnover")
plt.title("Turnover ot two Companies")

plt.legend()

#show the  bar chart
plt.show()
```
